# Imports and Globals

```python
import numpy as np
import pandas as pd

INPUT_FILE = "Laliga.csv"
```

# Question: 1
Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (2.5 points)


```python
data = pd.read_csv(INPUT_FILE,skiprows=1)
data = data.replace("-",0)

data = data.apply(pd.to_numeric, errors='ignore')

print(data.dtypes)
data.shape
```

# Question: 2
Print all the teams which have started playing between 1930-1980. (5 points)

```python
print(data.loc[(data["Debut"] >= "1930") & (data["Debut"] <= "1980")].Team.values)
```

# Question: 3
Print the list of teams which came Top 5 in terms of points (2.5 points)

```python
data.sort_values(by=['Points'], ascending=False)[:5][["Team", "Points"]]
```

# Question: 4
Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)

Goal_diff_count = GoalsFor - GoalsAgainst

```python
def Goal_diff_count(df):
    return df["GoalsFor"]-df["GoalsAgainst"]

data["GoalDiff"] = Goal_diff_count(data)
print("Team: {} has the maximum goal difference of: {}".format(data.iloc[data["GoalDiff"].idxmax()]["Team"],
                                                               data.iloc[data["GoalDiff"].idxmax()]["GoalDiff"]))

print("Team: {} has the minium goal difference of: {}".format(data.iloc[data["GoalDiff"].idxmin()]["Team"],
                                                               data.iloc[data["GoalDiff"].idxmin()]["GoalDiff"]))
```

# Question: 5
Create a new column with name “Winning Percent” and append it to the data set (5 points)

Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

Print the top 5 teams which has the highest Winning percentage

```python
data["Winning Percent"] = (data["GamesWon"]/data["GamesPlayed"])*100
data["Winning Percent"] = data["Winning Percent"].fillna(0)
print(data["Winning Percent"])
```
